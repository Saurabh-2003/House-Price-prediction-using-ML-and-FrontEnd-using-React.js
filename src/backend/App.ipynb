## Now we reduce some features from dataset to reduce complexity of the dataset and only retain those we find useful

```python
X = df.drop(['date', 'id', 'price', 'waterfront', 'view', 'grade', 'sqft_above', 'zipcode', 'sqft_living15', 'sqft_lot15'], axis = 1)
y = df['price']
```

```python
X.describe()
```

## Using linear Regression on the dataset

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
# Splitting the dataset into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=2)
```

```python
lr.fit(X_train, y_train)
```

```python
#Calculating the r-score
lr.score(X_test, y_test)
```

```python
# Calculating the Mean_Squared_Error
from sklearn.metrics import mean_squared_error
y_pred = lr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
y_actual = pd.DataFrame(y_test).to_numpy().ravel()
df = pd.DataFrame({'Actual': y_actual, 'Predicted': y_pred})

# Plot the predicted vs actual values using Seaborn's scatterplot
sns.scatterplot(x='Actual', y='Predicted', data=df)

# Add a diagonal line for reference
plt.plot([df['Actual'].min(), df['Actual'].max()], [df['Actual'].min(), df['Actual'].max()], 'k--', lw=4)

# Set the title and axis labels
plt.title('Actual vs Predicted Prices')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')

# Show the plot
plt.show()
```

### we can see from above that the r2 score is low and mse score is very large that means our linear regression model is not performing well 

## Using Gradient Boosting Regression

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr = GradientBoostingRegressor(n_estimators = 100, learning_rate = 0.22, min_samples_split=2, max_depth=5, loss = 'squared_error')
```

```python
gbr.fit(X_train, y_train)
```

```python
gbr.score(X_test, y_test)
```

```python
# Calculating the Mean_Squared_Error
from sklearn.metrics import mean_squared_error
y_pred = gbr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
y_actual = pd.DataFrame(y_test).to_numpy().ravel()
df = pd.DataFrame({'Actual': y_actual, 'Predicted': y_pred})

# Plot the predicted vs actual values using Seaborn's scatterplot
sns.scatterplot(x='Actual', y='Predicted', data=df)

# Add a diagonal line for reference
plt.plot([df['Actual'].min(), df['Actual'].max()], [df['Actual'].min(), df['Actual'].max()], 'k--', lw=4)

# Set the title and axis labels
plt.title('Actual vs Predicted Prices')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')

# Show the plot
plt.show()
```

### We can see from above that the r2 score is .8265.... which is significantly better than the linear regression and MSE is also more than half of it that is mean our Gradient Boosting Regressor is performing good on the Model. Now, we will explore some more model and will chose the best one out of it
